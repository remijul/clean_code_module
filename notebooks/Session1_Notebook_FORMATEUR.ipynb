```python
# SOLUTION : Réécriture de la fonction f()

def calculate_revenue_by_category(df, tax_rate=0.2):
    """
    Calcule le revenu total TTC par catégorie.
    
    Filtre les transactions avec montant positif, applique le taux de taxe,
    puis agrège par catégorie.
    
    Parameters
    ----------
    df : pd.DataFrame
        DataFrame contenant les colonnes 'amount' (montant HT) et 'category'.
    tax_rate : float, optional
        Taux de taxe à appliquer (default: 0.2 soit 20%).
    
    Returns
    -------
    pd.Series
        Revenu TTC par catégorie, trié par ordre décroissant.
    
    Examples
    --------
    >>> df = pd.DataFrame({'amount': [100, 200, -50], 'category': ['A', 'B', 'A']})
    >>> calculate_revenue_by_category(df, tax_rate=0.1)
    category
    B    220.0
    A    110.0
    Name: amount_ttc, dtype: float64
    """
    # Filtrer les transactions positives
    df_positive = df[df['amount'] > 0].copy()
    
    # Calculer le montant TTC
    df_positive['amount_ttc'] = df_positive['amount'] * (1 + tax_rate)
    
    # Agréger par catégorie
    revenue_by_category = df_positive.groupby('category')['amount_ttc'].sum()
    
    return revenue_by_category.sort_values(ascending=False)
```

```python
# SOLUTION : Réécriture de la fonction g()

def remove_duplicates(items):
    """
    Supprime les doublons d'une liste en préservant l'ordre.
    
    Parameters
    ----------
    items : list
        Liste pouvant contenir des doublons.
    
    Returns
    -------
    list
        Liste sans doublons, ordre préservé.
    """
    # Version pythonique en une ligne (Python 3.7+)
    # dict.fromkeys() préserve l'ordre d'insertion et élimine les doublons
    return list(dict.fromkeys(items))

# Test
print(remove_duplicates([1, 2, 2, 3, 1, 4]))  # [1, 2, 3, 4]
```

```python
# SOLUTION : Réécriture de la fonction h()
import re  # L'import doit être en haut du fichier !

def normalize_string(text):
    """
    Normalise une chaîne : minuscules et caractères alphanumériques uniquement.
    
    Utile pour créer des identifiants, des slugs ou comparer des chaînes
    indépendamment de la casse et de la ponctuation.
    
    Parameters
    ----------
    text : str
        Chaîne à normaliser.
    
    Returns
    -------
    str
        Chaîne en minuscules, sans caractères spéciaux.
    
    Examples
    --------
    >>> normalize_string('Hello World!')
    'helloworld'
    >>> normalize_string('Test-123_ABC')
    'test123abc'
    """
    text = text.lower()
    text = re.sub(r'[^a-z0-9]', '', text)
    return text

# Tests
print(normalize_string('Hello World!'))   # 'helloworld'
print(normalize_string('Test-123_ABC'))   # 'test123abc'
```

### 4.2 Exercice : Créer un module utils.py

Nous allons créer un fichier `utils.py` contenant nos fonctions réécrites.

```python
# Ce code crée le fichier utils.py

utils_content = '''
"""
Utilitaires pour le traitement de données.

Ce module contient des fonctions de nettoyage et de transformation
couramment utilisées dans nos pipelines de données.
"""

import re
import pandas as pd


def calculate_revenue_by_category(df, amount_col='amount', category_col='category', tax_rate=0.2):
    """
    Calcule le revenu total TTC par catégorie.
    
    Parameters
    ----------
    df : pd.DataFrame
        DataFrame avec les colonnes de montant et catégorie.
    amount_col : str
        Nom de la colonne contenant les montants HT.
    category_col : str
        Nom de la colonne de catégorie.
    tax_rate : float
        Taux de taxe (default: 0.2 soit 20%).
    
    Returns
    -------
    pd.Series
        Revenu TTC par catégorie, trié par ordre décroissant.
    """
    df_valid = df[df[amount_col] > 0].copy()
    df_valid['amount_ttc'] = df_valid[amount_col] * (1 + tax_rate)
    revenue = df_valid.groupby(category_col)['amount_ttc'].sum()
    return revenue.sort_values(ascending=False)


def remove_duplicates(items):
    """
    Supprime les doublons d'une liste en préservant l'ordre.
    
    Parameters
    ----------
    items : list
        Liste pouvant contenir des doublons.
    
    Returns
    -------
    list
        Liste sans doublons.
    """
    return list(dict.fromkeys(items))


def normalize_string(text):
    """
    Normalise une chaîne : minuscules et caractères alphanumériques uniquement.
    
    Parameters
    ----------
    text : str
        Chaîne à normaliser.
    
    Returns
    -------
    str
        Chaîne normalisée.
    
    Examples
    --------
    >>> normalize_string("Hello World!")
    'helloworld'
    """
    text = text.lower()
    text = re.sub(r'[^a-z0-9]', '', text)
    return text
'''

# Sauvegarde du fichier
with open('utils.py', 'w') as f:
    f.write(utils_content)
    
print("✓ Fichier utils.py créé !")
```

```python
# SOLUTION : Fonction de validation d'email

def is_valid_email(email):
    """
    Vérifie si une chaîne ressemble à un email valide.
    
    Parameters
    ----------
    email : str
        Chaîne à vérifier.
    
    Returns
    -------
    bool
        True si le format est valide, False sinon.
    
    Examples
    --------
    >>> is_valid_email('user@example.com')
    True
    >>> is_valid_email('invalid-email')
    False
    """
    import re
    # Pattern pour validation basique d'email
    # - Commence par des caractères alphanumériques, points ou tirets
    # - Suivi de @
    # - Suivi d'un domaine avec au moins un point
    # - Se termine par une extension de 2+ caractères
    pattern = r'^[\w.-]+@[\w.-]+\.[a-zA-Z]{2,}$'
    return bool(re.match(pattern, str(email)))

# Tests
print(is_valid_email('user@example.com'))    # True
print(is_valid_email('user.name@domain.fr')) # True
print(is_valid_email('invalid'))             # False
print(is_valid_email('user@@domain.com'))    # False
print(is_valid_email(''))                    # False
```

```python
# SOLUTION COMPLÈTE

# Constantes nommées (évite les magic numbers)
MIN_VALUE_THRESHOLD = 100
MAX_RATIO_THRESHOLD = 1000


def calculate_average_ratio_by_type(
    df,
    value_col='value',
    size_col='size',
    type_col='type',
    min_value=MIN_VALUE_THRESHOLD,
    max_ratio=MAX_RATIO_THRESHOLD
):
    """
    Calcule le ratio valeur/taille moyen par type.
    
    Filtre les données selon des seuils de valeur minimale et de ratio maximum,
    puis calcule la moyenne du ratio par catégorie de type.
    
    Parameters
    ----------
    df : pd.DataFrame
        DataFrame source contenant les colonnes de valeur, taille et type.
    value_col : str, optional
        Nom de la colonne de valeur (default: 'value').
    size_col : str, optional
        Nom de la colonne de taille (default: 'size').
    type_col : str, optional
        Nom de la colonne de type/catégorie (default: 'type').
    min_value : float, optional
        Seuil minimum de valeur pour inclure une ligne (default: 100).
    max_ratio : float, optional
        Seuil maximum de ratio pour exclure les outliers (default: 1000).
    
    Returns
    -------
    pd.Series
        Ratio moyen par type, indexé par type_col.
    
    Raises
    ------
    ZeroDivisionError
        Évité en filtrant les lignes où size_col == 0.
    
    Examples
    --------
    >>> df = pd.DataFrame({
    ...     'value': [200, 300, 50, 400],
    ...     'size': [2, 3, 1, 4],
    ...     'type': ['A', 'A', 'B', 'B']
    ... })
    >>> calculate_average_ratio_by_type(df)
    type
    A    100.0
    B    100.0
    Name: ratio, dtype: float64
    """
    # Copie pour éviter de modifier l'original
    df_filtered = df.copy()
    
    # Filtrer les valeurs en dessous du seuil minimum
    df_filtered = df_filtered[df_filtered[value_col] > min_value]
    
    # Exclure les lignes avec taille nulle (évite division par zéro)
    df_filtered = df_filtered[df_filtered[size_col] != 0]
    
    # Calculer le ratio valeur/taille
    df_filtered['ratio'] = df_filtered[value_col] / df_filtered[size_col]
    
    # Exclure les ratios aberrants (outliers)
    df_filtered = df_filtered[df_filtered['ratio'] < max_ratio]
    
    # Calculer la moyenne par type
    average_ratio_by_type = df_filtered.groupby(type_col)['ratio'].mean()
    
    return average_ratio_by_type


# Test de la fonction
test_df = pd.DataFrame({
    'value': [200, 300, 50, 400, 150],
    'size': [2, 3, 1, 4, 0],  # Note: une ligne avec size=0
    'type': ['A', 'A', 'B', 'B', 'A']
})

result = calculate_average_ratio_by_type(test_df)
print(result)
```
